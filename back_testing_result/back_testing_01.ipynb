## 오직 음봉일때 매수후 매도 조건

```python
def can(date, ticker, profit):
    df = pybithumb.get_ohlcv(ticker)
    df = df[date]
    con = df['open'] < df['close']
    df['candle'] = np.where(con, 1, 0)
    df['h/o'] = np.round(df['high'] / df['open'], 4)
    df['l/o'] = np.round(df['low'] / df['open'], 4)
    df['c/o'] = np.round(df['close'] / df['open'], 4)

    
    con = df['h/o'] > profit
    df['result'] = np.where(con, profit, df['c/o'])
    df = df[df['l/o'] <= 0.88]
    con = df['l/o'] > 0.88
    df['result2'] = np.where(con, (df['c/o'] - 0.96)+1, 1-(0.96-df['c/o']) )
    return df
```

```python
dates = ['2019-01', '2019-02', '2019-03', '2019-04', '2019-05',
         '2019-06', '2019-07']
tickers = pybithumb.get_tickers()
new_df = pd.DataFrame()
for date in dates:
    for ticker in tickers:
        try:
            result = can(date, ticker, 1.02)['result'].cumprod()[-1]
            new_df = new_df.append({
                'date' : date,
                'ticker' : ticker,
                'result' : result
            }, ignore_index=True)
        except:
            pass
        print(new_df)
new_df = new_df[['date', 'ticker', 'result']]
new_df = new_df.sort_values(by=['date', 'result'], ascending=False)
```
