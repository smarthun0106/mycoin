## ma3 or ma5 상승장 + 전전일 거래량 대비 전일 거래량 30%이하 하락

```python
import pybithumb
import pandas as pd
import numpy as np
```

```python
def test01(date, ticker):
    df = pybithumb.get_ohlcv(ticker)
    df = df[date]
    df['ma5'] = df['close'].rolling(window=3).mean().shift(1)
    df['ma5/open'] = df['ma5']/df['open']
    df['bull_ma'] = (df['ma5/open'] < 1.02) & (df['ma5/open'] > 1.00)
    
    df['volume1'] = df['volume'].shift(1)
    df['volume2'] = df['volume'].shift(2)
    df['v1/v2'] = df['volume1'] / df['volume2']
    df['v_bull'] = df['v1/v2'] < 0.4
    
    fee = 0.0032
    df['ror'] = np.where(df['bull_ma'] & df['v_bull'],
                     df['close']/df['open'] - fee,
                     1)
    df['hpr'] = df['ror'].cumprod()
    return df
```

```python
test01('2019', 'LINK')
```
