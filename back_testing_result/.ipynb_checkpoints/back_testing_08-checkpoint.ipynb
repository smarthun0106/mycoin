```python
import pybithumb
import numpy as np
import math
```

```python
def test01(ticker, year):
    df = pybithumb.get_ohlcv(ticker)
    df = df[year].astype(int)
    df['gap'] = df['high'] / df['open']
    fee = 0.0032
    df['ror'] = np.where(df['gap'] > 1.005+fee, 1.01, df['close'] / df['open'])
    df['hpr'] = df['ror'].cumprod()
    return df['hpr'][-2]
```

```python
test01('TRX', '2019')
```
