```python
import pybithumb
import pandas as pd
import numpy as np
```

```python
def test01(date='2019', ticker='BTC', k=0.5):
    df = pybithumb.get_ohlcv(ticker)
    df = df[df.index > date]
    
    # 전일 ma3, 전일 ma8
    df['ma3'] = df['close'].rolling(window=3).mean().shift(1)
    df['ma8'] = df['close'].rolling(window=8).mean().shift(1)
    
    # range
    df['range'] = (df['high'].shift(1) - df['low'].shift(1)) * k
    df['target'] = df['range'] + df['open']
    
    # deco
    df.dropna(inplace=True)
    df = df.astype(int)
    
    # make ma bull
    df['ma_bull'] = (df['ma3'] > df['open']) & (df['ma8'] > df['open'])
    
    # ror
    fee = 0.0032
    df['ror'] = np.where((df['high'] > df['target']) & df['ma_bull'],
                     (df['close'])/df['target'] - fee,
                     1)
    df['hpr'] = df['ror'].cumprod()
    return df
```

```python
test01(date='2018', ticker="LINK", k=0.5)
```
