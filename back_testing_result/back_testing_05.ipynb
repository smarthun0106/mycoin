## Noise + ma3이상

```python
import pybithumb
import time
import datetime
import pandas as pd
import numpy as np
```

```python
def test01(date, ticker):
    df = pybithumb.get_ohlcv(ticker)
    df = df[date]
    
    # ma3
    df['ma3_1'] = df['close'].rolling(window=3).mean().shift(1)
    df['open_1'] = df['open']
    df['open_1'] = df['open_1'].shift(1)
    df['open_1/ma3_1'] = df['open_1'] / df['ma3_1']
    df['bull'] = df['open_1/ma3_1'] > 0.95
    # target
    
    fee = 0.0032
    df['ror'] = np.where(df['bull'],
                     df['close']/df['open'] - fee,
                     1)
    df['hpr'] = df['ror'].cumprod()
    return df
```

```python
test01('2019', 'LINK')
```
