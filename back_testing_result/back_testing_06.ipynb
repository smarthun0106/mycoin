# Moving Average Cross Over Test 

```python
import pybithumb
import pandas as pd
import numpy as np
```

```python
df = pybithumb.get_ohlcv('BTC').astype(int)
df = df[2019]
```

```python
def add_ma(ticker='BTC', date='2019', low_ma=3, high_ma=8, k=1.005):
    df = pybithumb.get_ohlcv(ticker).astype(int)
    df = df[date]
    
    # add_low_ma
    low_ma_name = 'ma'+str(low_ma)
    df[low_ma_name] = df['close'].rolling(window=low_ma).mean()
    
    # add_high_ma
    high_ma_name = 'ma'+str(high_ma)
    df[high_ma_name] = df['close'].rolling(window=high_ma).mean()
    
    # deco
    df.dropna(inplace=True)
    df = df.astype(int)
    df['ticker'] = ticker
    
    # add candle
    df['candle'] = np.where(df['close']/df['open'] > 1.00, '양', '음')
    
    # ma5 > ma8_offset, 1, 0
#     condition = (df[high_ma_name+'_offset'] > df[low_ma_name]) & (df[high_ma_name+'_offset'] > 1.)
    df['cross_over'] = df[low_ma_name] /  df[high_ma_name]
    condition = (df['cross_over'] >= 1.00) & (df['cross_over'] <= k)
    df['aa'] = np.where(condition, 1, 0)
    
    df['candle-1'] = df['candle'].shift(1)
    df['candle-2'] = df['candle'].shift(2)
    
    df['result+1'] = (df['close'] / df['open']).shift(-1)
    df['result+2'] = (df['close'] / df['open']).shift(-2)
    df['result+3'] = (df['close'] / df['open']).shift(-3)
    return df
```

```python
df = add_ma(ticker='LINK', low_ma=3, high_ma=7, k=1.04)
```
