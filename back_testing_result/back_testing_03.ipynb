```python
''' 
이번 백테스팅은 noise가 낮은 코인들을 
가격 ma5상승장 + 거래량 ma5 상승장 + 변동성돌파
했을때의 수익률을 보겠다.
'''
```

```python
import pybithumb
import time
import datetime
import pandas as pd
import numpy as np
```

```python
# noise data
def noise():
    tickers = pybithumb.get_tickers()
    new_df = pd.DataFrame()
    for ticker in tickers:
        df = pybithumb.get_ohlcv(ticker)
        df = df[df.index > '2017']
        
        df['ticker'] = ticker
        
        df['range'] = (df['high'] - df['low']) * 0.5
        df['target'] = df['open'] + df['range'].shift(1)
        
#         df['noise'] = np.round(1 - (abs(df['open']-df['close']) / (df['high']-df['low'])), 2)
#         df['noise_ma20'] = df['noise'].rolling(window=20).mean().shift(1)
#         df['noise'] = df['noise'].shift(1)
        
        
        df['ma5'] = df['close'].rolling(window=5).mean().shift(1)
        df['bull_ma'] = df['open'] > df['ma5']
        
#         df['volume'] = df['volume']
#         df['v_ma5'] = df['volume'].rolling(window=5).mean().shift(1)
#         df['bull_vma'] = df['volume'] > df['v_ma5']
        

        new_df = new_df.append(df)
    return new_df
```

```python
noise_df = noise()
noise_df
```

```python
def aaa(date):
    df = noise_df[date].sort_values(by='noise').head(2)
    fee = 0.0032
# ror

    df['ror'] = np.where(df['bull_ma'] & df['bull_vma'],
                         df['close']/df['target']-fee,
                         1)
    return df
```

```python
date = pd.date_range(start='2019-01-01', end='2019-07-15')
date
```

```python
aa = date.strftime('%Y-%m-%d')
```

```python
result_df = pd.DataFrame()
for i in aa:
    df = aaa(i)
    result_df = result_df.append(df)
```
